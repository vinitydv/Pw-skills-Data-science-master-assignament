```python
Q4. Why LookupError class is used? Explain with an example KeyError and IndexError.
```

```python
Ans : LookupError This is the base class for those exceptions that are raised when a key or index used on a mapping or sequence is invalid or not found. The exceptions raised are :

KeyError
IndexError
```

```python
try:
    l = [3,4,5,6,7]
    print(l[5])
except LookupError as le:
    print(IndexError)
```

```python
try:
    dic = {'key1':25,'key2':36}
    print(dic['key3'])
except LookupError as le:
    print(KeyError)
```

```python
Q5. Explain ImportError. What is ModuleNotFoundError?
```

```python
Ans : ImportError
An ImportError is raised when the import statement is unable to load a module or when the “from list” in from … import has a name that cannot be found.

ModuleNotFoundError
This is the subclass of ImportError which is raised by import when a module could not be found. It is also raised when None is found in sys.modules.
```
