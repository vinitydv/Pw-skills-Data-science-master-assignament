## Q3. What is the range of prices in the dataset? What is the minimum and maximum price?

## Q10. How does the rating of apps vary by category? Create a boxplot to compare the ratings of different app categories.

```python
import pandas as pd
import matplotlib.pyplot as plt

# Load the dataset
df = pd.read_csv('https://raw.githubusercontent.com/krishnaik06/playstore-Dataset/main/googleplaystore.csv')


# Select the columns we need
df = df[['Category', 'Rating']]

# Create a boxplot
plt.figure(figsize=(12,8))
plt.boxplot([df[df['Category'] == category]['Rating'].dropna() for category in df['Category'].unique()], labels=df['Category'].unique())
plt.xticks(rotation=90)
plt.title('App Ratings by Category')
plt.xlabel('Category')
plt.ylabel('Rating')
plt.show()
```

## Q12. What is the relationship between the size of an app and its rating? Create a scatter plot to visualize the relationship.

```python
import pandas as pd
import matplotlib.pyplot as plt

# Load the dataset
df = pd.read_csv('googleplaystore.csv')

# Remove rows with missing values in the Size and Rating columns
df.dropna(subset=['Size', 'Rating'], inplace=True)

# Convert the Size column to a numeric data type
df['Size'] = pd.to_numeric(df['Size'].apply(lambda x: x[:-1]))

# Create a scatter plot
plt.scatter(df['Size'], df['Rating'])

# Set the axis labels
plt.xlabel('Size (MB)')
plt.ylabel('Rating')

# Show the plot
plt.show()
```

## Q13. How does the type of app affect its price? Create a bar chart to compare average prices by app type.

```python
import pandas as pd
import matplotlib.pyplot as plt

# Load the Google Playstore dataset
google_playstore_df = pd.read_csv('https://raw.githubusercontent.com/krishnaik06/playstore-Dataset/main/googleplaystore.csv')

# Remove rows with missing values
google_playstore_df = google_playstore_df.dropna()

# Convert price column to float
google_playstore_df['Price'] = google_playstore_df['Price'].str.replace('$', '').astype(float)

# Group the data by app type and calculate the average price for each group
app_type_price = google_playstore_df.groupby('Type')['Price'].mean()

# Create a bar chart to compare average prices by app type
plt.bar(app_type_price.index, app_type_price.values)
plt.xlabel('App Type')
plt.ylabel('Average Price')
plt.title('Average Price by App Type')
plt.show()
```
