Q1: What is Matplotlib? Why is it used? Name five plots that can be plotted using the Pyplot module of
Matplotlib.

Ans :

Matplotlib is a python libraty that used to data visualization by using several types of grapg
It is used to read data in a grapg form
five plots in matplots
1. Line Plot
2. Scatter Plot
3. Bar Plot
4. Histogram
5. Box Plot


```python
import numpy as np
import matplotlib.pyplot as plt
np.random.seed(3)
x = 3 + np.random.normal(0,2,50)
y = 3 + np.random.normal(0,2,len(x))
plt.xlabel("X-Axis")
plt.ylabel("Y-Axis")
plt.title("This is title")
plt.scatter(x,y)
plt.show()
```

Q4: What is a bar plot? Why is it used? Using the following data plot a bar plot and a horizontal bar plot.

import numpy as np

company = np.array(["Apple", "Microsoft", "Google", "AMD"])

profit = np.array([3000, 8000, 1000, 10000])

Ans :

1. Bar plot is a plot that shows relationship between a numerical and a categorical variable.
2. Bar plots are used to compare different categories of data. They are useful for displaying data that is not continuous, such as nominal or ordinal data.
 

```python
import numpy as np
import matplotlib.pyplot as ply
company = np.array(["Apple", "Microsoft", "Google", "AMD"])
profit = np.array([3000, 8000, 1000, 10000])
plt.subplot(1,2,1)
plt.title("Bar Plot")
plt.bar(company,profit)


plt.subplot(1,2,2)
plt.title("Horizontal Bar Plot" )
plt.barh(company,profit)
```

Q5: What is a box plot? Why is it used? Using the following data plot a box plot.

box1 = np.random.normal(100, 10, 200)

box2 = np.random.normal(90, 20, 200)

Ans :

1. box plot use box to represent data on graph
2. box plot shows 50% of the data, with a line,
3. Indicating the median value
4. It is also shows data range

```python
import numpy as np
import matplotlib.pyplot as plt
box1 = np.random.normal(100, 10, 200)
box2 = np.random.normal(90, 20, 200)
plt.subplot(1,2,1)
plt.boxplot(box1)

plt.subplot(1,2,2)
plt.boxplot(box2)
plt.show()
```
